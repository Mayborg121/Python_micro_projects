1. You will be developing a simple program to simulate a bank account. The user is prompted to enter their initial balance. They can then choose to either deposit money, withdraw money, or check their balance. The program should keep running until the user decides to exit. Implement this using if-else, elif statements, and a while loop.(itâ€™s up to your logics)

```python
print("*** Welcome to IDFC Bank ***")
pin= "0000"
amount= 0
tries=3
while tries > 0:
    entered= input("Enter Your pin :")
    amount = int(input("Enter Initial Balance"))
    if entered == "0000" :
        stay="y"
        while stay == "y" :
            
            print('''Services :
                1. [Check Balance]: [1]
                2. [Withdraw cash]: [2]
                3. [Reset Pin]:     [3]
                4. [EXIT]:          [4]''')
            choice = input("Enter your choice : ")
            if choice == "1" :
                print("Your Account balance is : ",amount)
            elif choice == "2" :
                withdraw = int(input("Amount you want to withdraw : "))
                if withdraw>amount:
                    print("***Not Enough Balance***")
                else :
                    amount = amount - withdraw
                    print("Amount withdrawn successfully,")
                    print("Your Remaining Balance is : Rs." , amount)
            elif choice == "3" :
                attempts = 3
                while attempts>0 :
                    check=input("Enter your old PIN :")
                    if check == pin:
                        pin=input("Enter new PIN.")
                        print("PIN Reset Successful !!!")
                        break
                    else :
                        print("** Wrong PIN **")
                        print("You have",attempts,"remaining")
                        attempts=attempts-1
            elif choice == "4" :
                stay="n"
            else :
                print("** Invalid Choice **")
                
            stay=input("Do you want to stay [y/n] :")
            
    if entered == "n":
        break
    else :
        tries=tries-1
        print("Wrong Password!! , You have",tries ," attempts Left")
print("******  Thank you, have a good day !  ******")

```
