## Pandas .Series() with Python Lists[ ]

```python
lst = [1,2,3,4,5]
print(lst)
```

```python
series = pd.Series(lst)
print(series)
print(type(series))
```

```python
#creating empty serries
emty = pd.Series([])
print(emty)
```

```python
a = pd.Series(['a','b','c','d','e'], index = [10,11,12,13,14], name = "Alphabets")
print(a)
```

```python
#scalar series
scalar_a = pd.Series(0.7)
print(scalar_a)
scalar_a = pd.Series(0.7 , index = [1,2,3,4])
print(scalar_a)
```

## Pandas Series() with Python dictionary()

```python
dict_series = pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5})
print(dict_series)
print(dict_series[0]) #prints value at 0th index
print(dict_series[2:4]) #prints the rows from 2nd to 4th
max(dict_series) # prints the maximum of the values
```

```python
#list as values in series
b = pd.Series({'a':[1,2,3],'b':[4,5,6],'c':[7,8,9],'d':[1,1,2]})
print(b)
```

## Handling NULL / Missing Values

```python
import pandas as pd
df = pd.read_csv("C://users/MAYBORG/Downloads/sample.csv")
df
```

```python
df.isnull()
```

```python
df.isnull().sum() #sum of all null values in particular columns
```

```python
df.isnull().sum().sum()
```

```python
df2 = df.dropna() # we can use axis in the parameter , default = 0 i.e. rows
df2
```

```python
# Dropping the columns with the NaN values 
df3 = df.dropna(axis = 1) # axis =1 for columns 
df3
```

```python
df3.shape
```

```python
# if any row has null value remove the row 
df.dropna(how = 'any')
```

```python
# id all row values is null then remove the row.
df.dropna(how = 'all')
```

```python
df.dropna(inplace = True) #replaces the original dataframe
df.shape
```
